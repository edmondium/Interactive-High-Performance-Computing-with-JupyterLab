```r
options(repr.plot.width = 6)
options(repr.plot.height = 5)

t = seq(0, 2*pi, by = 0.01)
x = 16 * sin(t)^3
y = 13 * cos(t) - 5*cos(2*t) - 2*cos(3*t) - cos(4*t)
plot(x, y, main = "Heart with R", xlab = "", ylab = "", col = "red")
```
