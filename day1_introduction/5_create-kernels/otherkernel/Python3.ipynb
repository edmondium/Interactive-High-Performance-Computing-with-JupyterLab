```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize = [8, 7])
t = np.arange(0,2*np.pi, 0.1)
x = 16*np.sin(t)**3
y = 13*np.cos(t)-5*np.cos(2*t)-2*np.cos(3*t)-np.cos(4*t)
plt.plot(x,y)
plt.title("Heart with Python")
```
