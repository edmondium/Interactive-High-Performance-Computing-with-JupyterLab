### Customizing Keyboard Shortcuts
Shortcuts can be customized in JupyterLabs settings system, but not all are editable by the user. [more](https://jupyterlab.readthedocs.io/en/2.2.x/developer/extension_points.html?highlight=shortcut#keyboard-shortcuts)

**Internal details**  
The shortcut handler propagates up the DOM tree from the focused element and tests each element against the registered selectors. If a match is found, then that command is executed.

#### Exercise 10:
Add a shortcut for "Restart Kernel and Clear All Outputs"
- Open "Commands Panel" - `(ESC)` , `CTRL + SHIFT + C`
- Search for "Restart Kernel and Clear All Outputs" -> no keyboard shortcut available for "Restart Kernel and Clear All Outputs"
  - **!! Let´s add the missing shortcut !!**
- Advanced Settings Editor - `(ESC)` , `CTRL + ,`
  - Switch to JSON Settings Editor and select "Keyboard Shortcuts"
  - Attention: "User Preferences" are added **after** "System Defaults" => they have the higher priority
    - list of dictionary objects

- Advanced Settings Editor -> Keyboard Shortcuts
  - Find `CTRL + F` for "restart"
  - Find next `CTRL + G` until "restart-and-clear" is found
  - Add to "User Preferences"
```json
{
    "shortcuts": [
        {
            "command": "kernelmenu:restart-and-clear",
            "keys": [
                "R",
                "J"
            ],
            "selector": "[data-jp-kernel-user]:focus"
        },
        {
            "command": "runmenu:restart-and-run-all",
            "keys": [
                "E",
                "E"
            ],
            "selector": "[data-jp-kernel-user]:focus"
        }
   ]
}
```
<u>Note</u>: To disable a system default shortcut, copy it to User Preferences and add the "disabled" key 
```
{
    "shortcuts": [
        {
           "command": "application:activate-next-tab",
           "keys": [
               "Ctrl Shift ]"
           ],
           "selector": "body",
           "disabled": true
       }
   ]
}   
```

* Advanced Settings Editor -> Keyboard Shortcuts
    * "selector" - Requierd focus for the shortcut to take place.
    * syntax error check, no save possible

### Summary

**A selection of keyboard shortcuts**

| Function                   | Keyboard Shortcut         | Menu                 |
| :-                         | :-                        | :-                   |
| Run selected cell          | `CTRL + Enter`            | Run → Run Selected Cell and Don't Advance |
| Run selected cell and advance to next | `SHIFT + Enter` | Run → Run Selected Cell |
| Run selected cell and insert a new cell below | `ALT + Enter` | Run → Run Selected Cell and Insert Below |
| Split cell                 | `CTRL + SHIFT + -`        | Edit → Split Cell    |
| Merge selected cells       | `SHIFT + M`               | Edit → Merge Selected Cells |
| Undo cell operation        | `(ESC)` , `Z`             | Edit → Undo Cell Operation |
| Create new cell above      | `(ESC)`, `A`              | n/a |
| Create new cell below      | `(ESC)` , `B`             | n/a |
| Delete selected cells      | `(ESC)` , `D` , `D`       | Edit → Delete Cells |
| Undo cell operation        | `(ESC)` , `Z`             | Edit → Undo Cell Operation |
| Redo cell operation        | `(ESC)` , `SHIFT + Z`     | Edit → Redo Cell Operation |
| Change cell type to "Markdown" | `(ESC)` , `M`         | n/a |
| Change cell type back to "Code" | `(ESC)`, `Y`         | n/a |
| Copy cells                 | `(ESC)` , `C`             | Edit → Copy Cells |
| Paste cells below          | `(ESC)` , `V`             | Edit → Paste Cells Below |
| Save Notebook              | `(ESC)` , `CTRL + S`      | File → Save Notebook |
| Find                       | `CTRL + F`                | Edit → Find |
| Find next                  | `CTRL + G`                | Edit → Find Next |
| Find previous              | `CTRL + SHIFT + G`        | Edit → Find Previous |
| Restart kernel             | `(ESC)` , `0 , 0`         | Kernel → Restart Kernel |
